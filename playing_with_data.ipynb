# Find the Human Labels

```python
import json

imsitu = json.load(open("face-classification/imSitu/imsitu_space.json"))

nouns = imsitu['nouns']
```

```python
human_nouns = {}
for noun in nouns.keys():
    if any([x in ['human', 'man', 'woman', 'child', 'girl', 'boy', 'person'] for x in nouns[noun]['gloss']]):
        human_nouns[noun] = nouns[noun]
```

```python
noun_keys = human_nouns.keys()
```

# Get the image labels

```python
train = json.load(open('face-classification/imsitu/train.json'))
dev = json.load(open('face-classification/imsitu/dev.json'))
test = json.load(open('face-classification/imsitu/test.json'))

labels = dict(train)
labels.update(dev)
labels.update(test)
```

```python
human_images = {}
for label in labels.keys():
    frames = labels[label]['frames']
    if 'agent' in frames[0].keys():
        if any([x['agent'] in noun_keys for x in frames]):
            human_images[label] = labels[label]
    if 'individuals' in frames[0].keys():
        if any([x['individuals'] in noun_keys for x in frames]):
            human_images[label] = labels[label]
```
